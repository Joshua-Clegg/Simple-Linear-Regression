```python
# import the required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# load the dataframe and print the first five rows
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# create variables x and y that hold age and charge values
# create scatterplot
x=df['age'].values.reshape(-1,1)
y = df['charges']
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charge')
plt.show()
```

```python
# fit data to linear regression model
charge_model = LinearRegression()
charge_model.fit(x,y)
```

```python
y_pred =charge_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.ylabel('Charges')
plt.xlabel('Age')
plt.show()
```

According to this model, the line of best fit indicates the x value is directly proportional to the y value. 
This means that as age increases the charges tend to increase proportionally.

```python
# test a random x value to prove relationship
unk_x = [[33]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = charge_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print("Those aged 33 will be charged:", charge_model.predict(unk_x))
```
